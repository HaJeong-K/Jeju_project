## 통합

```python
re = pd.merge(tourinfo_2021, placecharm_2021, on=['분류', '세부분류'])
```

```python
re1 = pd.merge(re, trafficinfo_2021, on=['분류', '세부분류'])
re1.head(2)
```

```python
re2 = pd.merge(re1, conv_2021, on=['분류', '세부분류'])
re2.head(2)
```

```python
re3 = pd.merge(re2, shopping_2021, on=['분류', '세부분류'])
re3.head(2)
```

```python
re4 = pd.merge(re3, accom_2021, on=['분류', '세부분류'])
re4.head(2)
```

```python
re5 = pd.merge(re4, budget_2021, on=['분류', '세부분류'])
re5.head(2)
```

```python
re6 = pd.merge(re5, food_2021, on=['분류', '세부분류'])
re6.head(2)
```
