#### 1인당 지출 경비 - 관광, 문화 지출 비용
- 통계자료를 가져오기 전, 2018년 1월 기준 관광지별 요금에 대한 파일을 먼저 첨부해 당시 요금에 대해 언급한다.

```python
import chardet
import pandas as pd

DATA_path = "/content/drive/MyDrive/세프/DATA/"
input_filename = DATA_path + '관광지입장요금표(2018.01.01).csv'

# 파일의 인코딩을 감지
with open(input_filename, 'rb') as f:
    result = chardet.detect(f.read())

# 감지된 인코딩 출력
detected_encoding = result['encoding']
print(f"Detected Encoding: {detected_encoding}")

# 파일을 읽고 새로운 인코딩으로 저장
df = pd.read_csv(input_filename, encoding=detected_encoding)
output_filename = DATA_path + '관광지입장요금표_2018.csv'
df.to_csv(output_filename, encoding='utf-8', index=False)

print(f"File converted to 'utf-8' and saved as {output_filename}")
```

```python
import chardet
import pandas as pd

DATA_path = "/content/drive/MyDrive/세프/DATA/"
input_filename = DATA_path + '관광지입장요금표(2019.01.09).csv'

# 파일의 인코딩을 감지
with open(input_filename, 'rb') as f:
    result = chardet.detect(f.read())

# 감지된 인코딩 출력
detected_encoding = result['encoding']
print(f"Detected Encoding: {detected_encoding}")

# 파일을 읽고 새로운 인코딩으로 저장
df = pd.read_csv(input_filename, encoding=detected_encoding)
output_filename = DATA_path + '관광지입장요금표_2019.csv'
df.to_csv(output_filename, encoding='utf-8', index=False)

print(f"File converted to 'utf-8' and saved as {output_filename}")
```

#### 1인당 지출 경비 - 관광, 문화 지출 비용
- 통계자료를 가져오기 전, 2020년 1월 기준 관광지별 요금에 대한 파일을 먼저 첨부해 당시 요금에 대해 언급한다.

```python
import chardet
import pandas as pd

DATA_path = "/content/drive/MyDrive/세프/DATA/"
input_filename = DATA_path + '관광지입장요금표(2020.1.1).csv'

# 파일의 인코딩을 감지
with open(input_filename, 'rb') as f:
    result = chardet.detect(f.read())

# 감지된 인코딩 출력
detected_encoding = result['encoding']
print(f"Detected Encoding: {detected_encoding}")

# 파일을 읽고 새로운 인코딩으로 저장
df = pd.read_csv(input_filename, encoding=detected_encoding)
output_filename = DATA_path + '관광지입장요금표_2020.csv'
df.to_csv(output_filename, encoding='utf-8', index=False)

print(f"File converted to 'utf-8' and saved as {output_filename}")
```

  + 이 방법을 선택한 이유는 비교해야할 결측치가 겹치고, 구간을 정해 해당 부분을 채우려해도 연속된 몇달간의 데이터가 없기 때문에 활용에 문제가 생겨 다른 방법보다 의미있게 결측치를 대치할 수 있을 것 같아서 활용했다.

```python
import pandas as pd
import numpy as np
from scipy.interpolate import CubicSpline

months_to_add = ['03월', '04월', '05월', '06월', '07월']
to_concat = []

for month in months_to_add:
    if month not in total_cost_2020['세부 카테고리'].values:
        new_row = {'분석 카테고리': '월별', '세부 카테고리': month, '평균': np.nan}
        to_concat.append(pd.DataFrame([new_row]))

total_cost_2020 = pd.concat([total_cost_2020] + to_concat, ignore_index=True)
total_cost_2020 = total_cost_2020.sort_values(by='세부 카테고리')
total_cost_2020 = total_cost_2020.drop(columns=['사례수', '10만원 미만', '10만원 이상-30만원 미만', '30만원 이상-50만원 미만', '50만원 이상-100만원 미만', '100만원 이상-150만원 미만', '150만원 이상'])
total_cost_2020['평균'] = pd.to_numeric(total_cost_2020['평균'], errors='coerce', downcast='integer')
total_cost_2020 = total_cost_2020.reset_index(drop=True)

# Spline Interpolation
x = total_cost_2020.index.values
y = total_cost_2020['평균'].values
mask = ~np.isnan(y)
cs = CubicSpline(x[mask], y[mask])
interpolated_values = cs(x)
total_cost_2020['평균'] = interpolated_values
total_cost_2020['평균'] = np.round(total_cost_2020['평균']).astype(int)

total_cost_2020
```
