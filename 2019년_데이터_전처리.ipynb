### 데이터 수집 비율
- 2019년도 제주 여행자를 대상으로 한 방문 관광객 실태조사에서 응답자 프로파일을 가져옴.
  + 응답자 프로파일의 카테고리 중 월별, 연령별 카테고리만 수집함.
- **해당 연도의 데이터 수집 대상 비율을 명시해 앞으로 다룰 데이터들이 어떤 방식으로 추출되었는지 설명함.**

```python
import pandas as pd

DATA_path = '/content/drive/MyDrive/세프/DATA/'

profile_df = pd.read_excel(DATA_path + '2019년 응답자 프로파일')
profile_df.head()
```

```python
profile_df = profile_df.drop(index = 0)
profile_df.head()
```

```python
columns = {'table' : '분석 카테고리',
           'Unnamed: 1' : '세부 카테고리',
           'Unnamed: 2' : '사례수',
           'Unnamed: 3' : '비율(%)'}

profile_df = profile_df.rename(columns = columns)
profile_df
```

```python
profile_df = profile_df.drop(index=range(14, 30)).reset_index(drop=True)

profile_df
```

- 전처리가 완료된 응답자 프로파일.
  + 월별, 연령별로 나눠진 사례들의 비율이 모두 비슷하게 퍼져있음을 확인할 수 있다.
  + 단, 61세 이상 부분에만 층화추출을 적용한 것으로 확인된다.

```python
profile_df = profile_df.drop(index=range(19, 29)).reset_index(drop=True)

profile_df
```
