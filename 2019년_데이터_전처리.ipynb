- 2019년도 제주 여행자를 대상으로 한 방문 관광객 실태조사에서 응답자 프로파일을 가져옴.

```python
import pandas as pd

DATA_path = '/content/drive/MyDrive/세프/DATA/'

profile_df_2019 = pd.read_excel(DATA_path + '2018년 응답자 프로파일 - 데이터 설명용.xlsx')
profile_df_2019.head()
```

- 전처리가 완료된 응답자 프로파일.
