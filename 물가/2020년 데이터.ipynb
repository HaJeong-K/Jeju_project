## 1인당 지출 경비 전처리
- 1인당 지출 경비에 대한 카테고리: 전체 부분, 숙박 부분, 식음료 부분, 항공/선박 부분, 관광/문화 지출 부분
- 1인당 지출 경비 부분이 지출 기준인원 항목의 하위 항목으로 분류되어 있어 응답자 프로파일처럼 해당 데이터의 비율을 명시함.

```python
expenditure_standard_2020 = pd.read_excel('/content/drive/MyDrive/2020 지출 인원.xlsx')
expenditure_standard_2020.head()
```

```python
columns = {'table' : '분석 카테고리',
           'Unnamed: 1' : '세부 카테고리',
           'Unnamed: 2' : '사례수',
           'Unnamed: 3' : '1명',
           'Unnamed: 4' : '2명',
           'Unnamed: 5' : '3명',
           'Unnamed: 6' : '4명',
           'Unnamed: 7' : '5명',
           'Unnamed: 8' : '6명',
           'Unnamed: 9' : '7명',
           'Unnamed: 10' : '8명',
           'Unnamed: 11' : '9명',
           'Unnamed: 12' : '10명',
           'Unnamed: 13' : '11명 이상',
           'Unnamed: 14' : '평균'}

expenditure_standard_2020= expenditure_standard_2020.rename(columns = columns)
expenditure_standard_2020
```

```python
expenditure_standard_2020 =expenditure_standard_2020.drop(index = range(11, 46)).reset_index(drop=True)

expenditure_standard_2020
```

1) 월별 패턴:

- 1월과 2월에는 11명 이상의 그룹이 비교적 많이 나타나고, 9월과 10월에는 2명에서 5명까지의 그룹이 더 많음.

2) 평균

- 전체 데이터에 대한 평균값을 보면 "01-02월"의 경우 약 3.17이고, "08-12월"의 경우 약 2.91입니다. 이를 통해 "01-02월"에는 더 많은 인원 수가 있는 경향이 있음.

```python
expenditure_standard_2020  = expenditure_standard_2020.drop(index = [0, 1]).reset_index(drop=True)

expenditure_standard_2020
```
