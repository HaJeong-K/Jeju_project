#### 1인당 지출 경비 - 여름시즌 관광지 이용 비용
- 통계자료를 분석하기 전, 2022년 1월 기준 관광지별 요금에 대한 파일을 먼저 첨부해 당시 요금에 대해 언급한다.

```python
import chardet
import pandas as pd

DATA_path = "/content/drive/MyDrive/세프/DATA/"
input_filename = DATA_path + '관광지입장요금표(2022.1.1).csv'

# 파일의 인코딩을 감지
with open(input_filename, 'rb') as f:
    result = chardet.detect(f.read())

# 감지된 인코딩 출력
detected_encoding = result['encoding']
print(f"Detected Encoding: {detected_encoding}")

# 파일을 읽고 새로운 인코딩으로 저장
df = pd.read_csv(input_filename, encoding=detected_encoding)
output_filename = DATA_path + '관광지입장요금표_2022.csv'
df.to_csv(output_filename, encoding='utf-8', index=False)

print(f"File converted to 'utf-8' and saved as {output_filename}")
```

- 전처리가 완료된 2022년도 관광지별 요금 (단체요금 X)
  + 2022년도 요금확인 용.
