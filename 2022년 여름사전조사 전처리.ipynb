### 제주여행비율
- 성수기로 꼽히는 7, 8월 중 어느 달을 여행달로 선택했는지

```python
DATA_PATH = '/content/drive/MyDrive/Colab Notebooks/jeju/'
plan_before = pd.read_excel(DATA_PATH + '2022_summer_bf/07-1 제주여행계획.xlsx',
                            skiprows=1,
                            skipfooter=1) # 첫 행과 마지막 행은 의미가 없어서 삭제
```

```python
columns = {'Unnamed: 0' : '분류',
           'Unnamed: 1' : '세부분류',
           '7월' : '7월선택률(%)',
           '8월' : '8월선택률(%)'}
plan_before = plan_before.rename(columns = columns)
plan_before
```

### 선정 시 고려했던 국가/국내지역

```python
consider_before = pd.read_excel(DATA_PATH + '2022_summer_bf/07-22 선정시 고려한 국가국내지역.xlsx',
                            skiprows=1,
                            skipfooter=1)
consider_before.head()
```

```python
columns = {'Unnamed: 0' : '분류',
           'Unnamed: 1' : '세부분류'}
consider_before = consider_before.rename(columns = columns)
consider_before.head()
```

```python
consider_before.set_index(consider_before.columns[0], inplace=True)
new_index_name='분류'
drop_rows = ['권역', '제주여행 계획월', '직업', '거주지역']
consider_before1= consider_before.drop(drop_rows)
consider_before1
```

- 당시 코로나 여파로 인해 해외여행 제약이 아직 심했던 터라 해외여행 검토율이 낮게 나타남
- 가구소득이 낮을수록 제주도를 선택한 비율이 높게 나타남

### 내년 여행 계획

```python
nextyear_before = pd.read_excel(DATA_PATH + '2022_summer_bf/07-23 내년 여행 계획(해외여행 자유화시).xlsx',
                                skiprows=1,
                                skipfooter=1)
nextyear_before.head()
```

```python
columns = {'Unnamed: 0' : '분류',
           'Unnamed: 1' : '세부분류',
           '내년까지는 제주도를 우선적으로 여행할 것이다' : '제주도 우선',
           '제주도와 해외여행을 모두 갈 것이다' : '모두 함',
           '제주도 대신 해외여행을 우선적으로 할 것이다' : '해외여행 우선',
           '제주도와 해외여행을 모두 하지 않을 것이다' : '모두 하지 않음',
           '잘 모르겠다' : '모름'}
nextyear_before = nextyear_before.rename(columns = columns)
nextyear_before.head()
```

```python
new_col_order = ['분류', '세부분류', '사례수', '제주도 우선', '해외여행 우선', '모두 함', '모두 하지 않음', '모름']
nextyear_before = nextyear_before[new_col_order]
nextyear_before.head()
```

```python
nextyear_before.set_index(nextyear_before.columns[0], inplace=True)
new_index_name='분류'
drop_rows = ['권역', '제주여행 계획월', '직업', '거주지역']
nextyear_before1= nextyear_before.drop(drop_rows)
nextyear_before1
```

- 사전계획과 비슷하게 내년 계획의 경우도 소득에 비례해 해외여행&제주도와 해외를 모두 고려하는 편이 높게 나타남

### 국내 관광지 대비 물가 인식 수준

```python
price_aw_before = pd.read_excel(DATA_PATH + '2022_summer_bf/10-01 국내 관광지 대비 물가인식수준.xlsx',
                            skiprows=1,
                            skipfooter=1)
price_aw_before.head()
```

```python
columns = {'Unnamed: 0' : '분류',
           'Unnamed: 1' : '세부분류',
           '②<-------' : '별로 비싸지 않다',
           '③비슷하다' : '비슷하다',
           '④------->' : '조금 비싸다',
           '⑤매우 비싸다' : '매우 비싸다',
           'BOT2(①+②)' : '비싸지 않다',
           'MID(③)' : '비슷하다',
           'TOP2(④+⑤)' : '비싸다'}
price_aw_before = price_aw_before.rename(columns = columns)
price_aw_before.head()
```

```python
price_aw_before.set_index(price_aw_before.columns[0], inplace=True)
new_index_name='분류'
drop_rows = ['권역', '제주여행 계획월', '직업', '거주지역']
price_aw_before1= price_aw_before.drop(drop_rows)
price_aw_before1
```

- 5점 척도일텐데 1점은 데이터가 없는지 아예 누락되어 있음(bottom 데이터가 1과 2를 포함하기에 설문지에 존재는 하는 것으로 보임)
- 조금비싸다, 매우비싸다를 합친 비싼편으로 보는 의견이 압도적인 편. 소득을 가리지 않음.
- 15-20세가 수치가 조금 낮게 나온 편인데 미성년의 경우 여행 비용을 직접 벌고 쓰는 편이 아니라 체감을 상대적으로 못하는 것이 아닐까 하는 개인적인 의견
