# 추측
- 귀무가설 : 성수기와 비성수기 간의 각 항목별 만족 100점 평균은 유의미한 차이가 없다
- 대립가설 : 성수기와 비성수기 간의 각 항목별 만족 100점 평균은 유의미한 차이가 있다
- 성수기 : 7~8월, 비성수기 : 나머지 월
- 100점 평균으로 t-test진행
- column들이 항목이 아니라 척도와 평균점수이기 때문에 요인분석이 불가하다고 판단
- 데이터 양이 많지 않아 alpha를 0.1로 두고 측정
- 타 데이터(만족 항목 등)로도 다른 검정을 시도해보려 했는데, 성수기와 비성수기간 데이터양이 많이 차이가 나서 계속 코드 오류가 발생함

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
summer_data = result_jeju[(result_jeju['세부분류'] == '07월') | (result_jeju['세부분류'] == '08월')]
other_months_data = result_jeju[(result_jeju['세부분류'] != '07월') & (result_jeju['세부분류'] != '08월')]
```

## 관광정보 서비스에 대한 만족도 차이

```python
summer_info_sco = summer_data['관광정보 100점평균']
other_months_info_sco = other_months_data['관광정보 100점평균']
```

```python
t_statistic_info, p_value_info = ttest_ind(summer_info_sco, other_months_info_sco)
```

```python
alpha = 0.10  # 유의수준 설정
if p_value_info < alpha:
    print("귀무 가설을 기각합니다. 성수기와 비성수기 간 관광 정보 100점 평균에는 유의미한 차이가 있습니다.")
else:
    print("귀무 가설을 기각하지 않습니다. 성수기와 비성수기 간 관광 정보 100점 평균에는 유의미한 차이가 없습니다.")

print('관광정보 t-통계량 :', t_statistic_info)
print('관광정보 p-값 :', p_value_info)
```

```python
plt.figure(figsize=(10, 5))

# 박스 플롯
plt.subplot(1, 2, 1)
infob = plt.boxplot([summer_info_sco, other_months_info_sco],  patch_artist=True)
plt.xticks([1, 2], ['성수기', '비성수기'])
plt.title('관광 정보 만족 100점 평균 - 박스 플롯')
infob['boxes'][0].set(facecolor='red')
infob['boxes'][1].set(facecolor='blue')
for median in infob['medians']:
    median.set(color='black', linewidth=2)
plt.tight_layout()
plt.show()
```
