```python
info_2018['연도']=2018
info_2019['연도']=2019
info_2020['연도']=2020
info_2021['연도']=2021
```

```python
merge_info= pd.concat([info_2018, info_2019, info_2020, info_2021], axis=0)
merge_info
```

```python
from sklearn.preprocessing import StandardScaler
feature = ['자국의 인터넷 사이트/앱',	'한국의 여행관련 사이트/앱',	'자국 여행사(오프라인)',	'관광안내책자',	'친지, 친구, 동료',	'현지 한국 공공기관 및 박람회',	'과거여행경험',	'주요 언론매체(TV, 라디오, 신문, 잡지 등)',	'항공사, 호텔',	'기타']
x = merge_info.loc[:, feature].values
y = merge_info.loc[:, ['세부 카테고리']].values
x_transformed = StandardScaler().fit_transform(x)
x_transformed.shape
```

```python
from factor_analyzer import FactorAnalyzer
fa = FactorAnalyzer(n_factors=2, method="ml", rotation=None)

fa.fit(x)
```

```python
efa_result=pd.DataFrame(fa.loadings_, index=feature)
print(efa_result)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'NanumGothic'

plt.figure(figsize = (6,10))
sns.heatmap(efa_result, cmap="Blues", annot =True, fmt='.2f')
```

```python
from sklearn.preprocessing import StandardScaler
feature = ['여행정보를 얻기 어렵다',	'관광정보가 정확하지 않다',	'관광종사원이 불친절하다',	'안내표지판이 부정확하다',	'식당과 음식이 불결하다',	'음식이 입에 맞지 않는다',	'물가가 비싸다',	'대중교통 이용이 불편하다','택시기사 서비스가 불편하다',	'관광가이드의 서비스가 좋지 않다',	'쇼핑품목이 다양하지 않다',	'상품구입을 강요한다'	,'기타'	,'불만족하거나 불편했던 점이 없다']
x = merge_dis.loc[:, feature].values
y = merge_dis.loc[:, ['세부 카테고리']].values
x_transformed = StandardScaler().fit_transform(x)
x_transformed.shape
```
