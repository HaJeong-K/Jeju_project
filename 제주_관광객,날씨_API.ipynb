```python
weather = pd.read_excel('제주 강수량.xlsx')
```

```python
weather = weather.drop(index=range(0, 13)).reset_index(drop=True)
```

```python
column_names = {'Unnamed: 0': '지점번호',
                'Unnamed: 1': '지점명',
                'Unnamed: 2': '일시',
                'Unnamed: 3': '강수량(mm)',
                'Unnamed: 4': '일최다강수량(mm)',
                'Unnamed: 5': '일최다강수량일자',
                'Unnamed: 6': '1시간최다강수량(mm)',
                'Unnamed: 7': '1시간최다강수량일자'}
weather = weather.rename(columns=column_names)
weather
```

```python
weather = weather.drop(columns=['지점번호', '일최다강수량일자', '1시간최다강수량(mm)', '1시간최다강수량일자'])
weather
```

```python
weather['일시'] = weather['일시'].astype(str).str.replace(' 00:00:00', '')
weather['일시'] = weather['일시'].astype(str).str.replace('-01', '')
```

```python
weather.at[0, '일시'] = '2018-01'
weather.at[12, '일시'] = '2019-01'
weather.at[24, '일시'] = '2020-01'
weather.at[36, '일시'] = '2021-01'
weather.at[48, '일시'] = '2022-01'
weather.at[60, '일시'] = '2023-01'
weather
```

```python
weather = weather.rename(columns={'일시': '구분연월'})
weather
```

```python
column_names = {'목적별 - 교육여행': '목적별(교육여행)',
                '목적별 - 레저스포츠': '목적별(레저스포츠)',
                '목적별 - 기타방문': '목적별(기타방문)',
                '목적별 - 친지방문': '목적별(친지방문)',
                '목적별 - 회의및업무': '목적별(회의및업무)',
                '목적별 - 휴양및관람': '목적별(휴양및관람)',
                '행태별 - 개별여행': '행태별 (개별여행)',
                '행태별 - 부분패키지': '행태별(부분패키지)',
                '행태별 - 패키지': '행태별(패키지)',}

df_18 = df_18.rename(columns=column_names)
df_19 = df_19.rename(columns=column_names)
df_22 = df_22.rename(columns={'행태별(개별여행)': '행태별 (개별여행)'})

result = pd.concat([df_18, df_19, df_20, df_21, df_22], ignore_index=True)
result
```
