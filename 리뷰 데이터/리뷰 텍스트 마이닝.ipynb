```python
# okt 사용해보기 from konlpy

from konlpy.tag import Okt
okt=Okt()

# 형태소
print("okt.morphs : ", okt.morphs('안녕하세요. 저는 딥러닝 입니다.'))

# 명사
print("okt.nouns : ", okt.nouns('안녕하세요. 저는 딥러닝 입니다.'))

# 형태소 + 품사
print("okt.pos : ", okt.pos('안녕하세요. 저는 딥러닝 입니다.'))
```

```python
file_path = "/content/bija_review.txt"
try:
    with open(file_path, "r", encoding="utf-8") as file:
         bija_txt = file.read()
except :
  None
print(bija_txt)
```

```python
from konlpy.tag import Okt
from collections import Counter
import numpy as np

okt=Okt()
stop_words = ['곳','제주도','비자림','것','수']
# print("okt.nouns : ",okt.nouns(bija_txt))
bija=okt.nouns(bija_txt)
bija = [word for word in bija if word not in stop_words]
counts = Counter(bija)
tags = counts.most_common(100)
print(tags)
```
